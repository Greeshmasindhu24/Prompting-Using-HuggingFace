```python
# zero shot
prompt1="""Classify the following review as Positive, Negative, or Neutral:
"The product works fine, but not that great."
Answer: """
```

```python
# One Shot
prompt2="""Classify the following review as Positive, Negative, or Neutral:
Example:
Review: "The camera quality is amazing, and the battery lasts all day!"
Sentiment: Positive

Now classify this review:
Review: "The product works fine, but not that great"
Sentiment: """
```

```python
# Few Shot Prompting
prompt3="""Classify the following review as Positive, Negative, or Neutral:
Examples:
Review: "The camera quality is amazing, and the battery lasts all day!"
Sentiment: Positive

Review: "The delivery was late, and the product was damaged."
Sentiment: Negative

Review: "The product is okay. It does what it says, but nothing special and camera is not great."
Sentiment: Neutral

Now classify this review:
Review: "The product works fine, but not that great"
Sentiment: """
```

```python
# Chain of Thought
prompt4="""Let's analyze the sentiment of the following review step by step.

Review: "The product works fine, but the battery drains too quickly."

Step 1: Identify positive aspects.
- The phrase "The product works fine" suggests that the user acknowledges the product functions properly.

Step 2: Identify negative aspects.
- The phrase "but the battery drains too quickly" indicates dissatisfaction with battery life.

Step 3: Determine the overall sentiment.
- Although the product works fine, the battery issue is a significant downside.
- Since the negative aspect outweighs the positive, the sentiment is **Negative**.

Final Answer: Negative

Review: "The camera quality is amazing, and the battery lasts all day!"
Final Answer:
 """
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
inputs = tokenizer(prompt3, return_tensors="pt")
outputs = model.generate(**inputs)
result = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Predicted Sentiment:", result)
```
